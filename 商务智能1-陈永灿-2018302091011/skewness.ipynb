```python
#计算对称正态分布的Skewness的数值
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu, sigma , num_bins = 0, 1, 50
x = mu + sigma * np.random.randn(1000000)
# 正态分布的数据
n, bins, patches = plt.hist(x, num_bins, density=True, facecolor = 'blue', alpha = 0.5)
# 拟合曲线
y = norm.pdf(bins, mu, sigma)
data=pd.Series(x)
skewness=data.skew()
print("正态分布skewness值:"+str(skewness))
plt.plot(bins, y, 'r--')
plt.xlabel('Expectation')
plt.ylabel('Probability')
plt.title('histogram of normal distribution: $\mu = 0$, $\sigma=1$')
plt.subplots_adjust(left = 0.15)
plt.show()
```

```python
#计算正偏移的Skewness的数值
import pandas as pd
positive_skewed_distribution=[-2.4,-2.2,-2.2,-2.1,-2,-1,0.2,0.3,1,2]
data=pd.Series(positive_skewed_distribution)
skewness=data.skew()
print(positive_skewed_distribution)
print("skewness值:"+str(skewness))
```

```python
#计算负偏移的Skewness的数值
import pandas as pd
negative_skewed_distribution=[-2,-1,0,1.2,3.5,4,4,4.2,5,5.2]
data=pd.Series(negative_skewed_distribution)
skewness=data.skew()
print(negative_skewed_distribution)
print("skewness值:"+str(skewness))
```

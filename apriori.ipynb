```python
def load_data():
    with open("./data_set/data.txt", encoding='utf-8') as f:
        data = [i[:-1].split(',') for i in f.readlines()]
    return data


def is_apriori(ck_item, lksub1):
    for ck in ck_item:
        sub_ck = ck_item - frozenset([ck])
        if sub_ck not in lksub1:
            return False
    return True


def create_c1(data):
    c1 = set()
    for items in data:
        for item in items:
            item_set = frozenset([item])
            c1.add(item_set)
    return c1


def create_ck(lk, k):
    len_lk = len(lk)
    list_lk = list(lk)
    ck = set()
    for i in range(len_lk):
        for j in range(1, len_lk):
            l1 = list(list_lk[i])
            l2 = list(list_lk[j])
            l1.sort()
            l2.sort()
            if l1[0:k - 2] == l2[0:k - 2]:
                ck_item = list_lk[i] | list_lk[j]
                if is_apriori(ck_item, lk):
                    ck.add(ck_item)
    return ck


def create_lk_by_ck(data, ck, min_support, support_data):
    data_count = {}
    lk = set()
    for t in data:
        for item in ck:
            if item.issubset(t):
                if item not in data_count:
                    data_count[item] = 1
                else:
                    data_count[item] += 1
    t_num = float(len(data))
    for item in data_count:
        if (data_count[item] / t_num) >= float(min_support):
            lk.add(item)
            support_data[item] = data_count[item] / t_num
    return lk


def create_l(data, min_support, k):
    c1 = create_c1(data)
    support_data = {}
    L1 = create_lk_by_ck(data, c1, min_support, support_data)
    lk_sub = L1.copy()
    L = [lk_sub]
    for i in range(2, k + 1):
        Ci = create_ck(lk_sub, i)
        Li = create_lk_by_ck(data, Ci, min_support, support_data)
        lk_sub = Li.copy()
        L.append(lk_sub)
    return L, support_data


def create_big_rules(L, support_data, min_conf):
    big_rules = []
    sub_list = []
    for i in range(0, len(L)):
        for freq in L[i]:
            for sub in sub_list:
                if sub.issubset(freq):
                    conf = support_data[freq] / support_data[freq - sub]
                    big_rule = (freq - sub, sub, conf)
                    if conf >= float(min_conf) and big_rule not in big_rules:
                        big_rules.append(big_rule)
            sub_list.append(freq)
    return big_rules


if __name__ == "__main__":
    data = load_data()
    min_support = input("输入最小支持度：")
    min_conf = input("输入最小置信度：")
    L, support_data = create_l(data, min_support, k=3)
    big_rules = create_big_rules(L, support_data, min_conf)
    for Lk in L:
        lk_size = len(Lk)
        if lk_size > 1:
            print("=" * 80)
            print("频繁" + str(len(list(Lk)[0])) + "项集\t\t\t\t\t\t\t支持度")
            print("=" * 80)
            for freq_set in Lk:
                print(freq_set, support_data[freq_set])
    print("=" * 80)
    print("强关联规则\t\t\t\t\t\t\t置信度")
    print("=" * 80)
    for item in big_rules:
        print(item[0], "==>", item[1], "  ", item[2])
```

```python
import pandas as pd
import math

#读取数据
excel_data= pd.read_excel("./data_set/房地产1.xlsx")
arr = excel_data.values

#标称相异矩阵
nominal_matrix=[]
for i in range(len(arr)):
    difference=[]
    for j in range(len(arr)):
        match=0
        if(arr[i][0]==arr[j][0]):
            match+=1
        if(arr[i][1]==arr[j][1]):
            match+=1
        if(arr[i][2]==arr[j][2]):
            match+=1
        result=(3-match)/3.0
        difference.append(result)
    nominal_matrix.append(difference)

#数值相异矩阵
max_size=0
max_year=0
max_floor=0
max_price=0
max_bedroom=0
for i in range(len(arr)):
    for j in range(len(arr)):
        if abs(arr[i][4]-arr[j][4])>max_size:
            max_size=abs(arr[i][4]-arr[j][4])
        if abs(arr[i][5]-arr[j][5])>max_year:
            max_year=abs(arr[i][5]-arr[j][5])
        if abs(arr[i][6]-arr[j][6])>max_floor:
            max_floor=abs(arr[i][6]-arr[j][6])
        if abs(arr[i][7]-arr[j][7])>max_price:
            max_price=abs(arr[i][7]-arr[j][7])
        if abs(arr[i][8]-arr[j][8])>max_bedroom:
            max_bedroom=abs(arr[i][8]-arr[j][8])
numerical_matrix=[]
for i in range(len(arr)):
    difference=[]
    for j in range(len(arr)):
        distance=(abs((arr[i][4]-arr[j][4])/max_size)+abs((arr[i][5]-arr[j][5])/max_year)+abs((arr[i][6]-arr[j][6])/max_floor)+abs((arr[i][7]-arr[j][7])/max_price)+abs((arr[i][8]-arr[j][8])/max_bedroom))/5
        difference.append(distance)
    numerical_matrix.append(difference)

    
#序数相异矩阵
ordinal_matrix=[]
for i in range(len(arr)):
    difference=[]
    for j in range(len(arr)):
        distance=math.sqrt(math.pow(arr[i][3]-arr[j][3],2))
        difference.append(distance)
    ordinal_matrix.append(difference)

#混合类相似度矩阵
result=[]
for i in range(len(arr)):
    difference=[]
    for j in range(i+1):

        distance = (1-(nominal_matrix[i][j]*3+numerical_matrix[i][j]*5+ordinal_matrix[i][j])/9.0)
        difference.append(distance)
    result.append(difference)
    
#矩阵

pd.DataFrame(result)
```
